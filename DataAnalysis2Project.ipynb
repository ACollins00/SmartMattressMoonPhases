Feature 3: Make 3 matplotlib or seaborn (or another plotting library) visualizations to display your data.

Note: Below demonstrates one matplotlib visualization. Further down there will be MORE THAN 3 matplotlib visualizations. 

```python
import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import drange
import numpy as np

df = pd.read_csv("SmartMattressStats - Sheet1 (2).csv")
print("Contents in csv file:", df)


plt.figure(figsize=(12,8))

plt.plot(df.Night, df.TtoFallAsleep, 'r.-', label="Time to Fall Asleep")
plt.plot(df.Night, df.Restful, 'b.-', label="Restful")
plt.plot(df.Night, df.Restless, 'y.-', label="Restless")
plt.legend(['Time to Fall Asleep', 'Restfull', 'Restless'])
plt.title("Sleep Quality Stats")
plt.suptitle("Graph 1")
plt.xlabel("Night")
plt.ylabel("Minutes")

plt.xticks(rotation='vertical')

plt.tick_params(axis='x', which='major', labelsize=6)

plt.tight_layout()

plt.show()
```

In the graph below, there are only TWO aspects being shown in order to highlight the areas that appear to correlate. (As noted earlier, a new moon phase is set to a value of 200 in order to show on a graph.) Approximately two days before the moon enters a new phase, there is a sharp increase in heart rate variability. A few times the sharp increase happens a couple of days after the moon enters a new phase. 

```python
df = pd.read_csv('HealthStats - Sheet1 (1).csv')
df = pd.read_csv('SmartMattressStats - Sheet1 (2).csv')
print("Contents in csv file:", df)

df = pd.merge(data1, data2,
         on='Night',
         how='outer')


plt.figure(figsize=(12,8))

plt.plot(df.Night, df.HRV, 'b.-', label="HRV")
plt.plot(df.Night, df.Moon, 'y.-', label="Moon")
plt.legend(['HRV', 'Moon'])
plt.title("Heart Rate Variability and Moon Phase Change")
plt.suptitle("Graph 7")
plt.xlabel("Night")
plt.ylabel("Minutes")

plt.xticks(rotation='vertical')

plt.tick_params(axis='x', which='major', labelsize=6)

plt.tight_layout()

plt.show()
```
